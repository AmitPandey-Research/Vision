```python
# the error in the above image is due to the fact that cv2 reads as bgr whereas matplotlib follows rgb image style,
#so we need to convert the bgr images to rgb using cv2 convert color function
```

```python
cvtColorimg1 = cv2.cvtColor(img1, cv2.COLOR_BGR2RGB )
cvtColorimg2 = cv2.cvtColor(img2, cv2.COLOR_BGR2RGB )
cvtColorimg3 = cv2.cvtColor(img3, cv2.COLOR_BGR2RGB )
cvtColorimg4 = cv2.cvtColor(img4, cv2.COLOR_BGR2RGB )
```

```python
plt.figure(figsize=(8,8))

plt.subplot(2,2,1).imshow(cvtColorimg1)

plt.subplot(2,2,2).imshow(cvtColorimg2)

plt.subplot(2,2,3).imshow(cvtColorimg3)

plt.subplot(2,2,4).imshow(cvtColorimg4)
```

```python
# to convert any rgb image to grayscale, we use dot product and just display one channel. Also grayscale images are 30% red,60% green and 11% blue. R+G+B=1

import numpy as np
def rgb2gray(rgb):
    return np.dot(rgb[...,:3], [0.299, 0.587, 0.144])#dot product to convert image to 2 dimensional gray scale

gray_image = rgb2gray(img1)#passing image to the function
#cv2_imshow(gray_image)# display using cv2.

plt.imshow(gray_image, cmap = plt.get_cmap('gray'))# to explicitly tell that the image is grayscale and not rgb.
plt.axis('off')
plt.show()

```
